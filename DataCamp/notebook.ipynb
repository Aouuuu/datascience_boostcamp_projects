## Task 3

Describe the distribution of all of the sales. Your answer must include a visualization that shows the distribution.

The histogram depicts the distribution of sales within the 'petsup' dataset. The x-axis represents the sales values, divided into ten equal-sized bins, while the y-axis represents the count of occurrences within each bin

![image-2](image-2.png)

From the graph, we can observe that the bars are more concentrated towards the left side of the histogram, with a longer tail extending towards the right. This suggests that the distribution of sales is **right-skewed**. A right-skewed distribution indicates that there are relatively more occurrences of higher sales values, while fewer occurrences of lower sales values. This suggests that there may be a few instances of exceptionally high sales, while the majority of sales fall within a certain range.

Understanding the skewness of the distribution is crucial for businesses to make informed decisions. In this case, knowing that the sales distribution is right-skewed allows businesses to identify and capitalize on sales opportunities at higher price points or with higher-value products. It may also be beneficial to investigate the factors that contribute to higher sales values and leverage them to enhance sales performance further.
